```python
# Here I've imported necessary module.
import time
from random import choice  # This 'choice' word normally used to pick a random element those are given in our container "named as: keys"

password = input("Enter password: ")

# Here I've already defined characters. The code will crack the password only if these characters are present in your given password
keys = "QWERTYUIOPASDFGHJKLZXCVBVNM0123456789abcdefghijklmnopqrstuvwxyz-=_!@#$%^&*()"

# Here I'm calculating time and attempts taken to crack password
start_time = time.time()
attempts = 0
cracked = ""

# Here first i created outer loop which will iterate through each char. in password 
for i in password:

    # Here I've created inner loop, which tries guessing each allowed character that I've given in the "keys" container
    for k in keys:    
        time.sleep(0.01)  #I used this line to make the code 100x slower to crack password
        attempts += 1
        # This below "if" condition will check if the guessed character matches with the given password (character by character)
        if k == i:
            cracked += k  #This will fix the matched character and adds in the password
            break

# This will stop time as soon as the password is cracked.
end_time = time.time()

print("Password cracked:", cracked)
print("Attempts:", attempts)
print("Time taken:",end_time - start_time, "seconds")
```
